```python
#Import Python Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read Amazon Data
df = pd.read_csv("Amazon_Cell_Phones.csv")
```

```python
# #Drop Missing values and their rows
# df.dropna(inplace = True)
```

```python
# Grab the brand and rating datasets
phone_data = df[["brand","rating_out_of_5"]]
```

```python
#Show only the first 3117 rows because they do not have missing values in the rows column. 
df_final = phone_data.head(3117)
```

```python
df_final
```

```python
#Calculate the average ratings. 
average_ratings = df_final.groupby('brand')['rating_out_of_5'].mean()
```

```python
# plt.figure(figsize = (10, 6))
# plt.bar(average_ratings.index, average_ratings.values)
# plt.title("Average Ratings for Different Phone Models")
# plt.xlabels("Phone Model")
# plt.ylabel("Average Rating")
# plt.xticks(rotation = 45, ha = 'right')
# plt.show()

#sort the ratings in ascending order of ratings. 
top_brands = average_ratings.sort_values(ascending = False)

# Show the top 10 brands
top_5_brands = top_brands.head(10)

top_5_brands
```

```python
#Find the most expensive phone on Amazon. 
# filte rthe 
df_phone = df[["model_name","Price (Dollar)"]]
df_phone
```

```python
df_sort = df_phone.sort_values(by =["Price (Dollar)"], ascending = False)
df_sort
```

```python
df_top10 =  df_sort.head(10)
df_top10
plt.figsize
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(df_top10["model_name"], df_top10["Price (Dollar)"], color = 'orange')
plt.title('Top 10 expensive phones on Amazon in 2024')
plt.xlabel('Model Name')
plt.ylabel('price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
